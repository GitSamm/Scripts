```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages

matplotlib.rcParams['text.usetex'] = True
plt.style.use('dark_background')
```

```python
# Before we can determine whether this is useful infromation, we need to determine whether the data converged to a well enough value.

# First start with bringing in the data.

rot0 =   np.genfromtxt("Energy_00.dat")
rot45 =  np.genfromtxt("Energy_045.dat")
rot90 =  np.genfromtxt("Energy_090.dat")
rot135 = np.genfromtxt("Energy_135.dat")
rot180 = np.genfromtxt("Energy_180.dat")
rot225 = np.genfromtxt("Energy_225.dat")
rot270 = np.genfromtxt("Energy_270.dat")
rot315 = np.genfromtxt("Energy_315.dat")
rot360 = np.genfromtxt("Energy_360.dat")
```

```python

totalEnergy = rot0[:,1]
totalEnergyDropLast = totalEnergy[:-1]
totalEnergyDropFirst = totalEnergy[1:]
totalEnergyDiff = np.abs(totalEnergyDropLast - totalEnergyDropFirst)
totalEnergyDiffFrac = totalEnergyDiff / totalEnergyDropLast

coatEnergy = rot0[:,3]
coatEnergyDropLast = coatEnergy[:-1]
coatEnergyDropFirst = coatEnergy[1:]
coatEnergyDiff = np.abs(coatEnergyDropLast - coatEnergyDropFirst)
coatEnergyDiffFrac = coatEnergyDiff / coatEnergyDropLast

totalEnergy45 = rot45[:,1]
totalEnergyDropLast45 = totalEnergy45[:-1]
totalEnergyDropFirst45 = totalEnergy45[1:]
totalEnergyDiff45 = np.abs(totalEnergyDropLast45 - totalEnergyDropFirst45)
totalEnergyDiffFrac45 = totalEnergyDiff45 / totalEnergyDropLast45

coatEnergy45 = rot45[:,3]
coatEnergyDropLast45 = coatEnergy45[:-1]
coatEnergyDropFirst45 = coatEnergy45[1:]
coatEnergyDiff45 = np.abs(coatEnergyDropLast45 - coatEnergyDropFirst45)
coatEnergyDiffFrac45 = coatEnergyDiff45 / coatEnergyDropLast45

totalEnergy90 = rot90[:,1]
totalEnergyDropLast90 = totalEnergy90[:-1]
totalEnergyDropFirst90 = totalEnergy90[1:]
totalEnergyDiff90 = np.abs(totalEnergyDropLast90 - totalEnergyDropFirst90)
totalEnergyDiffFrac90 = totalEnergyDiff90 / totalEnergyDropLast90

coatEnergy90 = rot90[:,3]
coatEnergyDropLast90 = coatEnergy90[:-1]
coatEnergyDropFirst90 = coatEnergy90[1:]
coatEnergyDiff90 = np.abs(coatEnergyDropLast90 - coatEnergyDropFirst90)
coatEnergyDiffFrac90 = coatEnergyDiff90 / coatEnergyDropLast90

totalEnergy135 = rot135[:,1]
totalEnergyDropLast135 = totalEnergy135[:-1]
totalEnergyDropFirst135 = totalEnergy135[1:]
totalEnergyDiff135 = np.abs(totalEnergyDropLast135 - totalEnergyDropFirst135)
totalEnergyDiffFrac135 = totalEnergyDiff135 / totalEnergyDropLast135

coatEnergy135 = rot135[:,3]
coatEnergyDropLast135 = coatEnergy135[:-1]
coatEnergyDropFirst135 = coatEnergy135[1:]
coatEnergyDiff135 = np.abs(coatEnergyDropLast135 - coatEnergyDropFirst135)
coatEnergyDiffFrac135 = coatEnergyDiff135 / coatEnergyDropLast135

totalEnergy180 = rot180[:,1]
totalEnergyDropLast180 = totalEnergy180[:-1]
totalEnergyDropFirst180 = totalEnergy180[1:]
totalEnergyDiff180 = np.abs(totalEnergyDropLast180 - totalEnergyDropFirst180)
totalEnergyDiffFrac180 = totalEnergyDiff180 / totalEnergyDropLast180

coatEnergy180 = rot180[:,3]
coatEnergyDropLast180 = coatEnergy180[:-1]
coatEnergyDropFirst180 = coatEnergy180[1:]
coatEnergyDiff180 = np.abs(coatEnergyDropLast180 - coatEnergyDropFirst180)
coatEnergyDiffFrac180 = coatEnergyDiff180 / coatEnergyDropLast180

totalEnergy225 = rot225[:,1]
totalEnergyDropLast225 = totalEnergy225[:-1]
totalEnergyDropFirst225 = totalEnergy225[1:]
totalEnergyDiff225 = np.abs(totalEnergyDropLast225 - totalEnergyDropFirst225)
totalEnergyDiffFrac225 = totalEnergyDiff225 / totalEnergyDropLast225

coatEnergy225 = rot225[:,3]
coatEnergyDropLast225 = coatEnergy225[:-1]
coatEnergyDropFirst225 = coatEnergy225[1:]
coatEnergyDiff225 = np.abs(coatEnergyDropLast225 - coatEnergyDropFirst225)
coatEnergyDiffFrac225 = coatEnergyDiff225 / coatEnergyDropLast225

totalEnergy270 = rot270[:,1]
totalEnergyDropLast270 = totalEnergy270[:-1]
totalEnergyDropFirst270 = totalEnergy270[1:]
totalEnergyDiff270 = np.abs(totalEnergyDropLast270 - totalEnergyDropFirst270)
totalEnergyDiffFrac270 = totalEnergyDiff270 / totalEnergyDropLast270

coatEnergy270 = rot270[:,3]
coatEnergyDropLast270 = coatEnergy270[:-1]
coatEnergyDropFirst270 = coatEnergy270[1:]
coatEnergyDiff270 = np.abs(coatEnergyDropLast270 - coatEnergyDropFirst270)
coatEnergyDiffFrac270 = coatEnergyDiff270 / coatEnergyDropLast270

totalEnergy315 = rot315[:,1]
totalEnergyDropLast315 = totalEnergy315[:-1]
totalEnergyDropFirst315 = totalEnergy315[1:]
totalEnergyDiff315 = np.abs(totalEnergyDropLast315 - totalEnergyDropFirst315)
totalEnergyDiffFrac315 = totalEnergyDiff315 / totalEnergyDropLast315

coatEnergy315 = rot315[:,3]
coatEnergyDropLast315 = coatEnergy315[:-1]
coatEnergyDropFirst315 = coatEnergy315[1:]
coatEnergyDiff315 = np.abs(coatEnergyDropLast315 - coatEnergyDropFirst315)
coatEnergyDiffFrac315 = coatEnergyDiff315 / coatEnergyDropLast315

totalEnergy360 = rot360[:,1]
totalEnergyDropLast360 = totalEnergy360[:-1]
totalEnergyDropFirst360 = totalEnergy360[1:]
totalEnergyDiff360 = np.abs(totalEnergyDropLast360 - totalEnergyDropFirst360)
totalEnergyDiffFrac360 = totalEnergyDiff360 / totalEnergyDropLast360

coatEnergy360 = rot360[:,3]
coatEnergyDropLast360 = coatEnergy360[:-1]
coatEnergyDropFirst360 = coatEnergy360[1:]
coatEnergyDiff360 = np.abs(coatEnergyDropLast360 - coatEnergyDropFirst360)
coatEnergyDiffFrac360 = coatEnergyDiff360 / coatEnergyDropLast360


```

```python
# Load in the data that is going to be used to show the final value of the coating thermal noise.

data = np.genfromtxt("Energy.dat")
angle = data[:,0]
coatingthermalnoise = data[:,6]
```

```python
# Label sizes for the plots

myAxisLabelSize=32
myTickLabelSize=24
myLegendLabelSize=10
myTitleSize=25

gridwidth=0.5
gridcolor=(0.8,0.8,0.8)
```

```python
# Plot the Coating Energy for each rotation.

fig = plt.gcf()
ax = plt.gca()

gridwidth=0.5
gridcolor=(0.8,0.8,0.8)

ax.plot(rot0[:,0][:-1],   coatEnergyDiffFrac,    marker='o', label='$0^{\circ} Rotation$',   color='royalblue')
ax.plot(rot45[:,0][:-1],  coatEnergyDiffFrac45,  marker='o', label='$45^{\circ} Rotation$',  color='hotpink')
ax.plot(rot90[:,0][:-1],  coatEnergyDiffFrac90,  marker='o', label='$90^{\circ} Rotation$',  color='honeydew')
ax.plot(rot135[:,0][:-1], coatEnergyDiffFrac135, marker='o', label='$135^{\circ} Rotation$', color='cyan')
ax.plot(rot180[:,0][:-1], coatEnergyDiffFrac180, marker='o', label='$180^{\circ} Rotation$', color='indianred')
ax.plot(rot225[:,0][:-1], coatEnergyDiffFrac225, marker='o', label='$225^{\circ} Rotation$', color='teal')
ax.plot(rot270[:,0][:-1], coatEnergyDiffFrac270, marker='o', label='$270^{\circ} Rotation$', color='ivory')
ax.plot(rot315[:,0][:-1], coatEnergyDiffFrac315, marker='o', label='$315^{\circ} Rotation$', color='pink')
ax.plot(rot360[:,0][:-1], coatEnergyDiffFrac360, marker='o', label='$360^{\circ} Rotation$', color='limegreen')

ax.set_xlim([0,12])
ax.set_ylim([1e-4,1e0])
ax.set_yscale('log')

AxisLabelSize=20
ax.set_xlabel('$N$', fontsize=AxisLabelSize)
ax.set_ylabel('$|E_{N+1} - E_{N}|/ E_{N+1}$ ', fontsize=AxisLabelSize)

TickLabelSize=18
ax.tick_params(axis='x', labelsize=TickLabelSize, pad=12, length=1)
ax.tick_params(axis='y', labelsize=TickLabelSize, length=1)
ax.tick_params(which='minor', length=1)
ax.grid(b=True, which='major', color=gridcolor, linestyle='solid', linewidth=gridwidth)

ax.set_title('AlGaAs Coating Energy Convergence', fontsize=myTitleSize, color='w', y=1.04)
ax.legend(loc='lower left', fontsize=8)

fig.tight_layout()

plt.savefig('CrystalCoatingEnergyCon.png')

pdf_Convergence = PdfPages('CrystalCoatingEnergyCon.pdf')
pdf_Convergence.savefig()
pdf_Convergence.close()
```

```python
# Plot the Total Energy for each rotation.

fig = plt.gcf()
ax = plt.gca()

gridwidth=0.5
gridcolor=(0.8,0.8,0.8)

ax.plot(rot0[:,0][:-1],   totalEnergyDiffFrac,    marker='o', label='$0^{\circ} Rotation$',   color='royalblue')
ax.plot(rot45[:,0][:-1],  totalEnergyDiffFrac45,  marker='o', label='$45^{\circ} Rotation$',  color='hotpink')
ax.plot(rot90[:,0][:-1],  totalEnergyDiffFrac90,  marker='o', label='$90^{\circ} Rotation$',  color='honeydew')
ax.plot(rot135[:,0][:-1], totalEnergyDiffFrac135, marker='o', label='$135^{\circ} Rotation$', color='cyan')
ax.plot(rot180[:,0][:-1], totalEnergyDiffFrac180, marker='o', label='$180^{\circ} Rotation$', color='indianred')
ax.plot(rot225[:,0][:-1], totalEnergyDiffFrac225, marker='o', label='$225^{\circ} Rotation$', color='teal')
ax.plot(rot270[:,0][:-1], totalEnergyDiffFrac270, marker='o', label='$270^{\circ} Rotation$', color='ivory')
ax.plot(rot315[:,0][:-1], totalEnergyDiffFrac315, marker='o', label='$315^{\circ} Rotation$', color='pink')
ax.plot(rot360[:,0][:-1], totalEnergyDiffFrac360, marker='o', label='$360^{\circ} Rotation$', color='limegreen')

ax.set_xlim([0,12])
ax.set_ylim([5e-6,1e0])
ax.set_yscale('log')

AxisLabelSize=20
ax.set_xlabel('$N$', fontsize=AxisLabelSize)
ax.set_ylabel('$|E_{N+1} - E_{N}|/ E_{N+1}$ ', fontsize=AxisLabelSize)

TickLabelSize=18
ax.tick_params(axis='x', labelsize=TickLabelSize, pad=12, length=1)
ax.tick_params(axis='y', labelsize=TickLabelSize, length=1)
ax.tick_params(which='minor', length=1)
ax.grid(b=True, which='major', color=gridcolor, linestyle='solid', linewidth=gridwidth)

ax.set_title('AlGaAs Total Energy Convergence', fontsize=myTitleSize, color='w', y=1.04)
ax.legend(loc='lower left', fontsize=myLegendLabelSize)

fig.tight_layout()
plt.savefig('CrystalTotalEnergyCon.png')

pdf_TotalConvergence = PdfPages('CrystalTotalEnergyCon.pdf')
pdf_TotalConvergence.savefig()
pdf_TotalConvergence.close()
```

```python
fig = plt.gcf()
ax = plt.gca()

ax.plot(angle, coatingthermalnoise, marker='None', color='y')
ax.set_title('AlGaAs Coating Thermal Noise', fontsize=myTitleSize, color='w', y=1.04)

ax.set_xlabel('Angle', fontsize=myAxisLabelSize)
ax.set_ylabel(r'$\sqrt{S_q} \left(\mathrm{m}/\sqrt{\mathrm{Hz}}\right)$', fontsize=myAxisLabelSize)

ax.set_yscale('log')
plt.xticks(np.arange(min(angle), max(angle)+1, 90.0))

plt.grid(True)
ax.grid(b=True, which='minor', color=gridcolor, linestyle='solid', linewidth=gridwidth)

ax.tick_params(axis='x', labelsize=myTickLabelSize, pad=12, length=10)
ax.tick_params(which='minor', length=5)
ax.tick_params(axis='y', labelsize=myTickLabelSize, length=10)

fig.tight_layout()
plt.savefig('Rotation_CrystalAlGaAs_CoatingThermalNoise.png')

pdf = PdfPages('Rotation_CrystalAlGaAs_CoatingThermalNoise.pdf')
pdf.savefig()
pdf.close()
```
